<h1>Learning Seaborn</h1>

<p>Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Seaborn includes several built-in themes, color palettes, and functions for visualizing data distributions, relationships, and comparisons. Its syntax is simple and concise, making it easy to create complex plots with a few lines of code. Seaborn is widely used in data science, research, and other fields where data visualization is essential for understanding and communicating data insights.</p>


```python
import seaborn as sns
%matplotlib inline
```

```python
# Data Base to exemple
tips = sns.load_dataset('tips')
```

```python
tips
```

<h2>Distplot</h2>

<p>The distplot() function in Seaborn is used to plot a histogram with a line over it, representing the kernel density estimation (KDE) of the distribution. The kde parameter can be set to False to turn off the KDE curve. The bins parameter is used to specify the number of bins to be used in the histogram. The larger the number of bins, the more detailed the histogram, but it can also make it harder to interpret the data. By default, the distplot() function will use the Freedman-Diaconis rule to automatically determine the appropriate number of bins based on the size and distribution of the data.
</p>

```python
sns.distplot(tips['total_bill']);
```

```python
sns.distplot(tips['total_bill'], kde=False);
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=50);
```

<h2>Jointplot</h2>

<p>Jointplot() is a function from the Seaborn library used to plot bivariate data with different types of visualizations such as scatter plots, regression lines, and histograms. This function is useful to explore the relationship between two variables and identify patterns, correlations, and trends in the data. Additionally, it allows the user to add a parameter called "kind" to choose the type of plot to be shown. The default value is "scatter", but it is also possible to use "kde", "hex", "reg", and other options. </p>

```python
sns.jointplot(x='total_bill', y='tip', data=tips);
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg');
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex');
```

<h2>Pairplot()</h2>

<p>Pairplot() is a Seaborn function used to plot pairwise relationships between variables in a dataset. It creates a grid of scatterplots, histograms, and/or kernel density plots for each combination of the variables. This is a useful tool for exploring and visualizing the relationships between multiple variables at once. The Pairplot() function also allows for customization of the plots, such as choosing the type of plot to display and adding color coding based on a categorical variable.</p>

```python
sns.pairplot(tips);
```

```python
#The hue parameter in Pairplot allows grouping and visualizing data based on a third categorical variable.

sns.pairplot(tips, hue="sex");
```

```python
#The palette parameter of Pairplot allows to set the color palette used to represent different levels of the
#categorical variable specified in the hue parameter.

sns.pairplot(tips, hue="sex", palette="coolwarm");
```

<h2>rugplot()</h2>

<p>The rugplot() is a Seaborn function that plots a short line for each point in a one-dimensional dataset, allowing for a clear visualization of the data distribution. This function can be used in conjunction with other plotting functions to provide a more complete and accurate visualization of the data distribution.</p>

```python
sns.rugplot(tips['total_bill']);
```

```python
#Combining with kdeplot()

sns.kdeplot(tips['total_bill']);
```

```python
sns.kdeplot(tips['total_bill']);
sns.rugplot(tips['total_bill']);
```

<h2>Barplot()</h2>

<p>Barplot() is a function in Seaborn used to visualize categorical data using rectangular bars. It allows you to aggregate and display data using a statistical measure such as mean, median, or any other function. The height of each bar represents the value of the category being plotted, and the bars can be colored according to another categorical variable for better differentiation. Barplot() is a versatile function that can be customized with various parameters to create different types of bar charts for different data visualization needs.</p>

```python
sns.barplot(x='sex', y='total_bill', data=tips);
```

```python
import numpy as np
```

```python
#"Estimator" parameter in barplot function allows to define a function to aggregate the data before plotting,
# which provides a way to represent statistical estimates like mean, median, standard deviation, etc.

sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std);
```

<h2>countplo()</h2>

<p>The countplot() is a function of Seaborn that allows to plot the frequency of observations for a categorical variable. It is a useful tool for data exploration and allows to quickly identify the most common categories in a dataset. The countplot() function can also be customized with various parameters such as hue and palette to group and color the data according to a second categorical variable.</p>

```python
sns.countplot(x='sex', data=tips);
```

<h2>boxplot()</h2>

<p>The boxplot() is a function from Seaborn that creates a box-and-whisker plot to display the distribution of a dataset. It is particularly useful for comparing multiple groups of data and detecting outliers. The box in the plot represents the interquartile range (IQR), while the whiskers represent the rest of the distribution, excluding outliers. The function can also display additional information, such as the median and the range of the data. Overall, the boxplot() provides a simple and effective way to visualize the distribution of a dataset and to compare it across different groups or categories.
</p>

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow');
```

```python
#I swapped the x and y variables and used the parameter 'orient='h'' to present the graph horizontally.

sns.boxplot(y='day', x='total_bill', data=tips, palette='rainbow', orient='h');
```

```python
#The hue parameter in boxplot function allows to create boxplots comparing the distributions 
#of a numerical variable across different categories defined by a categorical variable.

sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='rainbow');
```

<h2>violinplot()</h2>

<p>The violinplot() is a Seaborn function used to visualize the distribution of numeric data across different levels of a categorical variable. It creates a combination of a boxplot and a kernel density plot, where the shape of the plot resembles a violin. The width of the violin at a particular point represents the density or number of data points present at that point. This plot provides information about the distribution of data as well as the quartiles, median, and outliers. It can also be used to compare the distribution of data between different categories by using the hue parameter. The violinplot() function is a useful tool for exploratory data analysis and data visualization.</p>

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow');
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, palette='rainbow');
```

```python
#The split parameter of the violinplot function can be used to compare the distributions 
#of two different groups side-by-side within the same violin plot.

sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, palette='rainbow', split=True);
```

<h2>lmplot()</h2>

<p>The lmplot() function in Seaborn is a powerful tool for creating regression plots that allow users to visualize the relationship between two variables and fit a regression model to the data.</p>

```python
sns.lmplot(x='total_bill', y='tip', data=tips);
```

```python
#We Can use the hue parameter

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex');
```

```python
#The markers parameter of lmplot() allows customization of the markers used in the
#scatter plot of the regression plot.

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['x','v']);
```

```python
#The scatter_kws parameter in lmplot() allows customization of the scatter plot elements, 
#such as marker size, edge color, and alpha.

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['x','v'], scatter_kws={'s': 100});
```

```python
"""The "col" parameter in lmplot() is used to create multiple regression plots, 
each for a specific category of the variable defined in "col". In this example, 
it creates separate plots for "Lunch" and "Dinner" in the "time" variable."""

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='time');
```

```python
"""O parÃ¢metro "row" no lmplot() Ã© usado para criar mÃºltiplos grÃ¡ficos de regressÃ£o em linhas 
separadas, cada uma para uma categoria especÃ­fica da variÃ¡vel definida em "row". Nesse exemplo,
ele cria grÃ¡ficos separados para cada dia da semana na variÃ¡vel "day", 
com diferentes grÃ¡ficos para "Lunch" e "Dinner" na variÃ¡vel "time"."""

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='time', row='day');
```

```python
#The "aspect" parameter sets the ratio of the width to the height of the regression plot.
#The "height" parameter defines the size of the regression plot.

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['x','v'], scatter_kws={'s': 100}, aspect=1.6, height=8);
```

<h2>Matrix Plots</h2>

<p>Seaborn is a Python data visualization library that offers several chart options, including matrix plots. These plots are extremely useful for visualizing the relationship between multiple variables simultaneously. With them, it is possible to generate a matrix of plots where each cell displays a scatter plot between two variables, a histogram, or a density plot. Additionally, it is possible to apply different colors and color maps to identify positive and negative correlations between variables. These plots are widely used in exploratory data analysis to identify patterns and trends among variables, as well as to conduct hypothesis tests. Seaborn offers several functions to generate different types of matrix plots, such as pairplot() and heatmap().</p>

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
#Transferring data to a matrix format using pivot_tables.

df_flights_matrix = flights.pivot_table(index='year', columns='month', values='passengers')
```

```python
df_flights_matrix
```

<h2>heatmap()</h2>

<p>The heatmap() function in Seaborn is a popular graphical tool for visualizing matrices of numeric data, where the color of each cell represents the magnitude of the value in the matrix. Heatmaps are particularly useful for identifying patterns and trends in large datasets and can be used to explore relationships between variables. With Seaborn's heatmap() function, you can create heatmaps that display the correlation between variables, plot the distribution of variables, or show the values of variables in a matrix. Additionally, you can customize the colors, labels, and annotations of the heatmap to highlight specific patterns or insights. Overall, the heatmap() function is a powerful tool for exploratory data analysis and data visualization.</p>

```python
sns.heatmap(df_flights_matrix);
```

```python
#one way to increase the size of the plot.

import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(20, 8))

sns.heatmap(df_flights_matrix, ax=ax);
```

```python
#The "annot" parameter in this "sns.heatmap()" function enables the annotation of each 
#cell in the heatmap with the respective value of the data.

fig, ax = plt.subplots(figsize=(20, 8))

sns.heatmap(df_flights_matrix, ax=ax, annot=True);
```

```python
#The parameter "fmt" in the function sns.heatmap() specifies the string format code for 
#annotating the heatmap cells, which in this case is set to display integers with no decimal places.

fig, ax = plt.subplots(figsize=(20, 8))
sns.heatmap(df_flights_matrix, ax=ax, annot=True, fmt='.0f');
```

```python
#The "linewidths" parameter in the sns.heatmap() function is used to set the 
#width of the lines that will separate each square in the heatmap.

fig, ax = plt.subplots(figsize=(20, 8))
sns.heatmap(df_flights_matrix, ax=ax, annot=True, fmt='.0f', linewidths=1);
```

```python
#The parameter "cmap" in the function sns.heatmap() sets the colormap used to map data 
#values to colors in the heatmap plot. In this example, it sets the colormap to "magma".

fig, ax = plt.subplots(figsize=(20, 8))
sns.heatmap(df_flights_matrix, ax=ax, annot=True, fmt='.0f', linewidths=1, cmap='magma');
```

<h2>clustermap</h2>

<p>The clustermap() function in Seaborn is a powerful tool for generating clustered heatmaps, which are useful for exploring patterns and similarities within complex datasets. This function works by first clustering both the rows and columns of the input data based on their similarities, and then generating a heatmap to visualize the resulting matrix. Additionally, the clustermap() function provides options to customize the clustering method, the color map, and the annotation of the heatmap with actual values. This function is particularly useful for visualizing large datasets with many variables, as it allows for quick identification of similar patterns and can help guide further analysis.</p>

```python
sns.clustermap(df_flights_matrix);
```

<h2>Styling the plots</h2>

<p>The set_style() function is a Seaborn method that allows users to customize the aesthetics of their plots. It allows them to set the overall style of the plot, such as background color, grid style, and font family, among others. Seaborn has five built-in themes: darkgrid, whitegrid, dark, white, and ticks, each with its own set of predefined style parameters. Users can also customize their own themes by using the set() method to modify specific style elements. This function is useful in creating plots that are not only informative but also visually appealing, which can help in better communicating insights from data to a broader audience.</p>

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips);
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips, palette='deep');
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips, palette='deep');
```

```python
sns.set_style('dark')
sns.countplot(x='sex', data=tips, palette='deep');
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips, palette='deep');
```
