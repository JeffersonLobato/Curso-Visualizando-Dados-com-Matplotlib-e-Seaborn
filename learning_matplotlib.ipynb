<h1>Learning Matplotlib</h1>

```python
import numpy as np

#The linspace method of numpy is used to create an array with equally 
#spaced values within a specified interval.

x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x ** 2
y
```

```python
#When importing matplotlib, conventionally we import it as plt.

import matplotlib.pyplot as plt

#The command "%matplotlib inline" is used to display matplotlib visualizations inline in Jupyter notebooks.
%matplotlib inline
```

```python
#The .plot() method is used in matplotlib to create a line plot or scatter plot of data.

plt.plot(x, y)
```

```python
#Some properties, such as line color, axis labels, and a title for the graph.

plt.plot(x, y, "red")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Title")
```

```python
#The plt.subplot(1, 2, 1) function is used in matplotlib to create a grid of subplots. 
#The parameters (1, 2, 1) specify that the grid has 1 row, 2 columns, 
#and the current subplot is the first one.

plt.subplot(1, 2, 1)

#The plt.plot(x, y, "r--") plots a line graph with red markers ("r") and dashed line style ("--").
plt.plot(x, y, "r--")

plt.subplot(1, 2, 2)

#The plt.plot(y, x, "g*--") plots a line graph with green star markers ("g*") and dashed line style ("--").
plt.plot(y, x, "g*--")
```

<h2>Object-oriented programming in matplotlib.</h2>

```python
"""In matplotlib, a figure is an object that contains all elements of a visualization, 
including plots, axes, labels, legends, and other text and image elements. When creating a figure, 
you can set its size and properties such as background color, title, and spacing between subplots. 
You can then add one or more subplots to the figure, each with its own set of axes and data. 
The figure is the foundation upon which you build your visualization, and it is common to save or 
display the figure once the visualization is complete."""

fig = plt.figure()


"""The fig.add_axes() method in matplotlib is used to add a new set of axes to a figure. 
It takes four parameters that specify the position and size of the new axes: the x position of
the bottom-left corner, the y position of the bottom-left corner, the width of the axes, and the 
height of the axes. For example, fig.add_axes([0.1, 0.1, 0.8, 0.8]) adds a new set of axes to the 
figure with origin point at (0.1, 0.1), width of 80% of the figure, and height of 80% of the figure.
These values can be adjusted to control the position and size of subplots within the figure."""

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, "blue")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, "blue")
axes.set_xlabel("Label x")
axes.set_ylabel("Label y")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, "blue")
axes.set_xlabel("Label x")
axes.set_ylabel("Label y")
axes.set_title("Title")

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(y, x, "red")
axes2.set_xlabel("Label x2")
axes2.set_ylabel("Label y2")
axes2.set_title("Title2")
```

<h2>Subplots()</h2>

```python
"""The subplots() function is used to create a grid of subplots in a figure. 
It returns an object containing an array of axes that can be used to modify each individual subplot.
It can be customized with various optional parameters to adjust the layout of the grid. 
It is an efficient way to create complex visualizations with multiple subplots in a single figure."""

fig, ax = plt.subplots()

ax.plot(x, y, "r--")
ax.set_title("Title of our subplot")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

for axis in ax:
    axis.plot(x, y, "r--")
    axis.set_title("Title of our subplot")
    

#The tight_layout() function is used to automatically adjust the spacing between 
#subplots so that they fit nicely in a figure.
plt.tight_layout()
```

```python
#The figsize=(20, 6) parameter of subplots() sets the dimensions of 
#the figure in which the grid of subplots will be plotted.

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))

ax[0].plot(x, y, "red")
ax[1].plot(x, y, "green")
```

<h2>Legends, labels and titles</h2>

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")

ax.set_title("Title")
ax.set_xlabel("Axis X")
ax.set_ylabel("Axis y")

ax.legend()
```
